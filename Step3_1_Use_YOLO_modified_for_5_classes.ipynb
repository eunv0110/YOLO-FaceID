#### 1) 데이터셋 불러오기

```python
import os
import zipfile
import random
```

```python
dataset1 = os.path.join('/content/drive/MyDrive/3조 이미지/수란/suran_face.zip') #
dataset2 = os.path.join('/content/drive/MyDrive/3조 이미지/예은/yeeun_face.zip')
dataset3 = os.path.join('/content/drive/MyDrive/3조 이미지/요한/yohan.zip')
dataset4 = os.path.join('/content/drive/MyDrive/3조 이미지/은비/eunbi_face.zip')
dataset5 = os.path.join('/content/drive/MyDrive/3조 이미지/종진/jongjin_face.zip' )
```

```python
# Colab에 생성할 본인 얼굴 폴더 경로
extract_folder = '/content/temp_dataset'

# 폴더가 없을 때 생성
if not os.path.exists(extract_folder):
    os.makedirs(extract_folder)

# 이미지와 라벨 폴더 생성
images_folder = os.path.join(extract_folder, 'images')
labels_folder = os.path.join(extract_folder, 'labels')

if not os.path.exists(images_folder):
    os.makedirs(images_folder)
if not os.path.exists(labels_folder):
    os.makedirs(labels_folder)

# 각 데이터셋에 대해 2500장씩 샘플링하여 추출
for dataset in [dataset1, dataset2, dataset3, dataset4, dataset5]:
    with zipfile.ZipFile(dataset, 'r') as zip_ref:
        # file_list에서 jpg 파일과 txt 파일만 구분
        jpg_files = [f for f in zip_ref.namelist() if f.lower().endswith('.jpg')]
        txt_files = [f for f in zip_ref.namelist() if f.lower().endswith('.txt')]

        # jpg 파일 2500장 무작위로 샘플링
        selected_jpg_files = random.sample(jpg_files, min(2500, len(jpg_files)))

        # 샘플링한 jpg 파일의 이름을 기준으로 해당 txt 파일 가져오기
        selected_txt_files = [f.replace('.jpg', '.txt') for f in selected_jpg_files if f.replace('.jpg', '.txt') in txt_files]

        # 샘플링한 파일을 해당 폴더로 복사
        for jpg_file in selected_jpg_files:
            with zip_ref.open(jpg_file) as source, open(os.path.join(images_folder, os.path.basename(jpg_file)), 'wb') as target:
                target.write(source.read())

        for txt_file in selected_txt_files:
            with zip_ref.open(txt_file) as source, open(os.path.join(labels_folder, os.path.basename(txt_file)), 'wb') as target:
                target.write(source.read())
```

#### 1) Label 파일의 클래스 값 변경

```python
import os
import glob
import random
import shutil
import numpy as np
```

```python
import glob
import os

# 각 데이터셋에 할당할 클래스 번호 지정
dataset_classes = {
    'suran_face': 0,
    'yeeun_face': 1,
    'yohan': 2,
    'eunbi_face': 3,
    'jongjin_face': 4
}

# txt 파일 전체 리스트화
txt_list = sorted(glob.glob(os.path.join('/content/temp_dataset/labels', '*.txt')))

for txt_file in txt_list:
    with open(txt_file, 'r') as f:
        lines = f.readlines()

    new_lines = []

    # txt 파일이 포함된 상위 폴더 이름 추출하여 클래스 번호 찾기
    dataset_folder = os.path.basename(os.path.dirname(os.path.dirname(txt_file)))
    class_id = dataset_classes.get(dataset_folder)

    # 클래스 번호가 설정되지 않은 경우 예외 처리
    if class_id is None:
        print(f"경고: {txt_file}에 대한 클래스 ID를 찾을 수 없습니다. 해당 파일을 건너뜁니다.")
        continue

    for line in lines:
        parts = line.strip().split()

        # 클래스 번호를 각 데이터셋의 고유 번호로 설정
        new_line = f'{class_id} ' + ' '.join(parts[1:])
        new_lines.append(new_line)
        print(f'클래스 {class_id}로 변경 : ', new_line)

    # 파일에 변경된 라벨 쓰기
    with open(txt_file, 'w') as f:
        f.write('\n'.join(new_lines) + '\n')

    print(f'파일 {txt_file} 처리 완료')
```

#### 2) 모델이 요하는 구조의 폴더 생성

```python
## 원본 폴더 경로들
source_folders = {'images': '/content/temp_dataset/images',
                  'labels': '/content/temp_dataset/labels'
                  }

## 타겟 폴더 경로들
target_folders = {'images':{'train':'/content/Datasets/images/train',
                            'val':'/content/Datasets/images/val',
                            # 'test':'/content/Datasets/images/test',
                            },
                  'labels':{'train':'/content/Datasets/labels/train',
                            'val':'/content/Datasets/labels/val',
                            # 'test':'/content/Datasets/labels/test',
                            }
                  }

## 폴더 생성 함수
def create_folder(path):
    if not os.path.exists(path):
        os.makedirs(path)

## Datasets 폴더 생성
create_folder('/content/Datasets')

## 하위 폴더 생성 : images, labels 및 train/val
for _, folders in target_folders.items() :
    for value in folders.values() :
        create_folder(value)
```

#### 3) 각 폴더에 데이터 옮기기

```python
## 데이터 스플릿 비율
split_size = 0.8

## images, labels 파일을 정렬하여 리스트화
image_list = sorted(glob.glob(os.path.join(source_folders['images'], '*')))
label_list = sorted(glob.glob(os.path.join(source_folders['labels'], '*')))

## 묶고 섞고 다시 리스트화
combined = list(zip(image_list, label_list))
random.shuffle(combined)
image_list, label_list = zip(*combined)

## 총 파일 수 계산
total_files = len(image_list)
tr_val_split = int(total_files * split_size)

print('전체 파일 수 : ', total_files)
print('training set 파일 수 : ', tr_val_split)
print('validation set 파일 수 : ', total_files - tr_val_split)

## train / val로 분리
tr_images, tr_labels = image_list[ : tr_val_split], label_list[ : tr_val_split]
val_images, val_labels = image_list[tr_val_split : ], label_list[tr_val_split : ]

print( 'training set 이미지 수 : ', len(tr_images), ' training set 라벨 수 : ', len(tr_labels) )
print( 'validation set 이미지 수 : ', len(val_images), 'validation set 라벨 수 : ', len(val_labels) )
```

```python
## 파일 이동 함수
def move_files(file_list, target_folder):
    for file_path in file_list:
        shutil.copy(file_path, os.path.join(target_folder, os.path.basename(file_path)))

# train, val로 이미지 및 라벨 파일 옮기기
move_files(tr_images, target_folders['images']['train'])
print("train set 이미지 파일 이동 완료!")
move_files(tr_labels, target_folders['labels']['train'])
print("train set 라벨 파일 이동 완료!")

move_files(val_images, target_folders['images']['val'])
print("val set 이미지 파일 이동 완료!")
move_files(val_labels, target_folders['labels']['val'])
print("val set 라벨 파일 이동 완료!")
```
